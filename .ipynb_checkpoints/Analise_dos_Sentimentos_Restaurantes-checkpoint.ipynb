```python
def clean_tripadvisor(tweet):
    try:
        tweet = BeautifulSoup(tweet, 'lxml').get_text() # o lxml indica o parse pode ser usado para dados de web por ex
        tweet = re.sub(r"\b\s-\s", ' ', tweet)
        # tweet = re.sub(r"\s....", '', tweet)
        tweet = re.sub(r"\b\.\s\\n", '. ', tweet)
        tweet = re.sub(r"\b\.", '. ', tweet)    
        return tweet
    except TypeError:
        print("Errou o tipo no regex")
text = 'O ambiente rústico e as comidas típicas do nordeste, desde o café regional a Buchada de Bode.Menu riquíssimo. Fiquei muito satisfeita, os funcionários todos de "Lampião e Maria Bonita", muito bom mesmo,'
print(clean_tripadvisor(text))
```

##Métodos de avaliação: Precisão, Renovação e Medida-F

```python
def whichIsSentenceFac(fac):
  if fac == 1.0:
    fac = 1
  elif fac <= 0.2:
    fac = 0
  elif fac > 0.2:
    fac = 1
  else:
    fac = 1
  return fac

_fac = whichIsSentenceFac(0.2)
print(_fac)

```

```python
def whichIsSentencePol(pol):
  if  pol == 0.0:
    pol = 0
  elif pol == 1.0:
    pol = 1
  elif pol > 0.0:
    pol = 1
  elif pol < 0.0:
    pol = -1
  elif pol == -1.0:
    pol = -1
  elif pol >= -0.026:
    pol = -1
  else:
    pol = 1
    
  return pol

_pol = whichIsSentencePol(-0.01)

print(_pol)
```

```python
def precision(take,total):
  prec = take/total
  precPercent = prec*100
  return precPercent
```

```python
def revogation(target,total):
  rev = target/total
  revPercent = rev*100
  return revPercent
```

```python
def medidaF1(prec,rev,all):
  _precision = precision(prec,all)
  _revogation = revogation(rev,all)
  resultF1 = (2*_precision*_revogation)/(_precision+_revogation)
  #print("resultado da medida F1 ",resultF1)
  return resultF1
```

###Fatiamento para o googletrans

```python
def splitToGoogleTrans(_comment):
  comment = clean_tripadvisor(_comment)
  sentenceGoogleTrans = []
  dot = '.'
  dotSpace = '. '

  if not dot or not dotSpace in comment:
    sentenceGoogleTrans.append(comment)
  if dot in comment:
    sentenceGoogleTrans = comment.split(dot)
  if dotSpace in comment:
    sentenceGoogleTrans = comment.split(dotSpace)
  else:
    print("alguma string diferente ",comment)

  _last = int(len(sentenceGoogleTrans))
  last = _last - 1
  
  if sentenceGoogleTrans[last] == '':
    sentenceGoogleTrans.pop(last)

  return sentenceGoogleTrans

splitToGoogleTrans('Pratos e entradas interessantes e gostosos. Os sanduíches achei sem graça (já experimentei dois diferentes). O ambiente é muito agradável, ainda mais se você gostar de música ao vivo à noite. Porém, acho os preços um pouco salgados. Bom para ocasiões especiais.')
   
```

###Verificação do idioma recebido

```python
def transGoogleCheckPt(sentences):
    translator = Translator()
    
    translationsPt = translator.detect([sentences])
    
    for lang in translationsPt:
        # print(lang.lang)
        pass
    test = lang.lang   
    if test == 'pt':
        translationsEn = translator.translate(sentences,origin='pt', dest='en')
        translationsEn = translationsEn.text
        translationsPt = sentences
        
    elif test == 'en':
        translationsEn = sentences
        translationsPt = translator.translate(translationsEn,origin='en', dest='pt')
        translationsPt = translationsPt.text
        
    else:
        translationsEn = translator.translate(sentences,origin='pt', dest='en')
        translationsEn = translationsPt.text
    return str(translationsEn), str(translationsPt)
```

###Detecção da língua portuguesa

```python
def portugueseSentence(out):
  saveResult = []
  saveResultToMetodo = []

  comentario = TextBlob(out)
  comentario_us = comentario.translate(to='us')

  print("Inglês: ",comentario_us)
  time.sleep(5)

  for sentence_pt, sentence in zip(comentario.sentences, comentario_us.sentences):    
    print('Setence->',sentence_pt, 'subj->', sentence.sentiment.subjectivity,'Pol->',sentence.sentiment.polarity)
    _fac = whichIsSentenceFac(float(sentence.sentiment.subjectivity))
    _pol = whichIsSentencePol(float(sentence.sentiment.polarity))
    resSubPol = [str(sentence_pt),_fac,_pol]
    saveResultToMetodo.append(resSubPol)
    #revBody, subj, pol = str(sentence_pt),float(sentence.sentiment.subjectivity),float(sentence.sentiment.polarity)
    temp = [str(sentence_pt),float(sentence.sentiment.subjectivity),float(sentence.sentiment.polarity)]
    saveResult.append(temp)

  return saveResult, saveResultToMetodo
```

###Inicio da rotina para reports

```python
def enterToTextBlob(out):
    try:
        whichLanguage = TextBlob(out)
        lang = whichLanguage.detect_language()
        if lang == 'en':
            print("It is english")
            _pt, _metodo = englishSentence(whichLanguage)

        elif lang == 'pt':
            print("It is portuguese")
            _pt, _metodo = portugueseSentence(out)
            #writeCsvTextBlob(_pt)
        else:
            print("Not english")
        return _pt , _metodo
    except TypeError:
        print("Errou não é tipo string")
```
